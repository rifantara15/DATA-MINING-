```python
X=digits.data
```

```python
X.shape
```

```python
sc = StandardScaler() 
X_std = sc.fit_transform(X)
```

```python
pca = decomposition.PCA(n_components=10) 
X_std_pca = pca.fit_transform(X_std) 
```

```python
X_std_pca.shape
```

```python
import pandas as pd 
import numpy as np 
```

```python
X_std_pca=pd.DataFrame(X_std_pca)
corr_matrix = X_std_pca.corr().abs() 
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), 
k=1).astype(np.bool)) 
 
to_drop = [column for column in upper.columns if any(upper[column] > 
0.95)] 
```

```python
X_std_pca.drop(X_std_pca.columns[to_drop], axis=1) 
```
